## Tâche 7 : diffractogramme du cristal $ScTlO_2$ pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
from pymatgen import Structure
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plot
```

```python
#téléchargement du fichier cif
with MPRester("jUGD1Q7NTzVGDpCRI27") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-755219")
    
    w = CifWriter(structure,symprec=1e-5);
    w.write_file('ScTlO2.cif')
```

```python
struct = Structure.from_file("ScTlO2.cif")
```

```python
XRD = XRDCalculator()
pattern = XRD.get_pattern(struct)
```

```python
print("Le premier pic du diffractogramme possède les caractéristiques suivantes :")
print("\n\tAngle 2\u03B8 :", pattern.x[0],"°")
print("\tIntensité relative :", pattern.y[0],"%")
print("\tIndices de Bravais-Miller associés :", pattern.hkls[0][0].get("hkl"))
print("\nLe deuxième pic du diffractogramme possède les caractéristiques suivantes :")
print("\n\tAngle 2\u03B8 :", pattern.x[1],"°")
print("\tIntensité relative :", pattern.y[1],"%")
print("\tIndices de Bravais-Miller associés :", pattern.hkls[1][0].get("hkl"))
print("\nLe troisième pic du diffractogramme possède les caractéristiques suivantes :")
print("\n\tAngle 2\u03B8 :", pattern.x[2],"°")
print("\tIntensité relative :", pattern.y[2],"%")
print("\tIndices de Bravais-Miller associés :", pattern.hkls[2][0].get("hkl"))
print("\nRemarque : on utilise ici le système de Bravais-Miller, qui ajoute un quatrième indice i aux indices de Miller (hkl) dans les réseaux hexagonaux et rhombohédrique.On note alors les indices (hkil), avec i tel que i+h+k=0. L'indice i n'apporte donc pas d'information mais permet de visualiser plus facilement les familles de plans.")
```

#### Ci-dessous, la figure du diffractogramme (limitée aux cinq premiers pics). Le cinquième pic est le pic d'intensité maximale.
